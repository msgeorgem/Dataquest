# Combined dete and tafe data using concat. And removed not necessary columns with null data

```python
combined_updated['institute_service'].value_counts()
```

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str')
```

```python
combined_updated['institute_service'].value_counts()
```

```python

replace_dict = {'Less than 1 year':'1','More than 20 years':'21','1-2':'1','3-4':'3','5-6':'5','11-20':'11','7-10':'7'}
combined_updated['institute_service'] = combined_updated['institute_service'].replace(replace_dict)

#combined_updated['institute_service'] = combined_updated['institute_service'].str.replace('.',' ')
def extract_last_word(element):
    last_word = str(element).split()[0]
    return last_word

combined_updated['institute_service'] = combined_updated['institute_service'].apply(extract_last_word)
#extracted = combined_updated['institute_service'].str.extract(pattern).astype(float)
combined_updated['institute_service'].value_counts()
```

```python
def mapping_years(value):
        #pd.isnull(value)
        if value =='nan':
            return np.nan
        else:
            value = float(value)
            if value < 3:
                return '1.New'
            elif (value >= 3 and value <= 6):
                return '2.Experienced'
            elif (value >= 7 and value <= 10):
                return '3.Established'
            else:
                return '4.Veteran'

combined_updated['service_cat'] = combined_updated['institute_service'].apply(mapping_years)
combined_updated['service_cat'].value_counts()
```
