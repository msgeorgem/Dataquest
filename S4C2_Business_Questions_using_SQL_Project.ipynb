```python
%%sql
WITH genre_counts AS (
SELECT g.name,ar.name artist, CAST(count(t.track_id) as float) counts
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
JOIN invoice_line il ON i.invoice_id = il.invoice_id
JOIN track t ON il.track_id = t.track_id
JOIN genre g ON t.genre_id = g.genre_id
JOIN album a ON t.album_id = a.album_id
JOIN artist ar ON a.artist_id = ar.artist_id
WHERE c.country = 'USA'
GROUP BY 1,2)
 
SELECT genre_counts.name, genre_counts.artist, CAST (genre_counts.counts as int) counts,
ROUND(CAST (genre_counts.counts/(
SELECT CAST (count(t.track_id) as float)total
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
JOIN invoice_line il ON i.invoice_id = il.invoice_id
JOIN track t ON il.track_id = t.track_id
JOIN genre g ON t.genre_id = g.genre_id
WHERE c.country = 'USA'
) as FLOAT)*100,2) percentage
FROM genre_counts
ORDER BY counts desc


```

# Write a query that finds the total dollar amount of sales assigned to each sales support agent within the company. Add any extra attributes for that employee that you find are relevant to the analysis

```python
%%sql
WITH genre_counts AS (
SELECT e.first_name||" "||e.last_name as full_name,
    c.company customer_name,c.city customer_city, ROUND (CAST(sum(i.total) as float),2) amount
FROM employee e
JOIN customer c ON e.employee_id = c.support_rep_id
JOIN invoice i ON c.customer_id = i.customer_id
GROUP BY 1,2,3)

select * from genre_counts
order by 4 desc
```

# Write a query that collates data on purchases from different countries.

    Where a country has only one customer, collect them into an "Other" group.
    The results should be sorted by the total sales from highest to lowest, with the "Other" group at the very bottom.
    For each country, include:
        total number of customers
        total value of sales
        average value of sales per customer
        average order value


```python
%%sql
WITH genre_counts AS (
SELECT c.country, count(c.customer_id), ROUND(sum(i.total),2), ROUND(avg(i.total),2)
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
JOIN invoice_line il ON i.invoice_id = il.invoice_id
GROUP BY 1)

select * from genre_counts
order by 2 desc
```

```python
%%sql
with invoice_count as (
    select invoice_id, album_id album_id_0, count(invoice_id) bought from  invoice_line il
JOIN track t ON il.track_id = t.track_id 
group by 1,2
), album_count as (select t.album_id album_id_1, count(t.track_id) exist from  invoice_line il
JOIN track t ON il.track_id = t.track_id 
group by 1)

select (select count(*) from 
invoice_count 
join album_count on invoice_count.album_id_0 = album_count.album_id_1
where invoice_count.bought = album_count.exist) full_albums, 

(select count(*) from 
invoice_count 
join album_count on invoice_count.album_id_0 = album_count.album_id_1
where invoice_count.bought != album_count.exist) singles, count(invoice_id)
from invoice 

```
