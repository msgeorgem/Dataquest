```python
import pandas as pd
import datetime as dt
import numpy as np

questions = pd.read_csv("2019_questions.csv", parse_dates=["CreationDate"])
questions.head()
```

```python
questions.info()
```

```python
questions["FavoriteCount"] = questions["FavoriteCount"].replace({np.NAN: 0})
questions["ViewCount"] = questions["ViewCount"].replace({np.NAN: 0})
questions["AnswerCount"] = questions["AnswerCount"].replace({np.NAN: 0})
```

```python
questions.info()
```

```python
questions = questions.dropna()
questions.info()
```

```python
questions["FavoriteCount"] = questions["FavoriteCount"].astype(int)
questions["ViewCount"] = questions["ViewCount"].astype(int)
questions["AnswerCount"] = questions["AnswerCount"].astype(int)
questions.info()
```

```python
questions.info()
```

## Transform tags column into distinct tags grouped and counted

```python
questions["Tags"]
```

```python
questions["Tags"]= questions["Tags"].str.replace('><', ',')
questions["Tags"]= questions["Tags"].str.replace('<', '')
questions["Tags"]= questions["Tags"].str.replace('>', '')
```

```python
questions["Tags"]
```

```python
from collections import Counter
r = Counter(",".join(questions["Tags"].values.tolist()).split(",")).items()
result = pd.DataFrame(list(r), columns=['Tag', 'Count'])
result

tagCount = result.sort_values(['Count'], ascending=[False])
tagCount
```

```python
wipViewCount = pd.concat([questions['ViewCount'], questions["Tags"].str.split(',', expand=True)], axis=1)
wipViewCount   
```

## Loop tag columns summing up viewcount  by every tag column and the result save in one dataframe 

```python
frames = list()
for item in range(0,5):
    group_by_tag = wipViewCount.groupby([item])
    count = group_by_tag['ViewCount'].agg('sum')
    countVies = pd.DataFrame(count)
    frames.append(countVies)

result = pd.concat(frames)
result['Tag'] = result.index
result.info()
result.reset_index(drop=True, inplace=True)
result
```

# result of grouping by tag columns group into distinct tags

```python
grouping = result.groupby('Tag')
count = grouping['ViewCount'].agg('sum')
countViewsFinal = pd.DataFrame(count)
countViewsFinal = countViewsFinal.sort_values(['ViewCount'], ascending=[False])
countViewsFinal['Tag1'] = countViewsFinal.index
##countViewsFinal = countViewsFinal.set_index('Tag1', inplace=True)
countViewsFinal
```

# Merge tagcounts with countviews per tag 

```python
res = pd.merge(tagCount, countViewsFinal, left_on='Tag', right_on='Tag1')
del res['Tag1']
res.head(20)
```
