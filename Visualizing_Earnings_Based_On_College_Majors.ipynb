```python
unique_major_category = recent_grads['Major_category'].unique()
major_women_share = {}
majors_predominantly_female = {}
majors_predominantly_male = {}

for item in unique_major_category:
    condition = cleaned_recent_grads['Major_category'] == item
    average_women_share = cleaned_recent_grads.loc[condition,'ShareWomen'].mean()
    
    major_women_share[item] = round(average_women_share,2)
    if average_women_share > 0.5:
        majors_predominantly_female[item] = round(average_women_share,2)
    else:
        majors_predominantly_male[item] = round(average_women_share,2)
        

major_women_share_series = pd.Series(major_women_share)
predominantly_male_series = pd.Series(majors_predominantly_male)
predominantly_female_series = pd.Series(majors_predominantly_female)
major_women_share_series
predominantly_male_series
predominantly_female_series
```

```python
df = pd.DataFrame(major_women_share_series, columns=['Major'])
df = pd.DataFrame(major_women_share_series, columns=['Women share'])
df
```
